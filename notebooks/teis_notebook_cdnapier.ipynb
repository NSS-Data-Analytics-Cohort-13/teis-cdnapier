# Create a function the buckets each kid into A-E for each domain, and run it on every domain and associated subdomains

```python
def get_category_2sub(exit_DQ, entry_DQ = 0, sub1_entry_rawScore=0, sub1_exit_rawScore=0, sub2_entry_rawScore=0, sub2_exit_rawScore=0, sub1_ZScore = -1.49, sub2_ZScore = -1.49):
    if (exit_DQ < 78) & (sub1_exit_rawScore <= sub1_entry_rawScore) & (sub2_exit_rawScore <= sub2_entry_rawScore):
        return 'a'
    elif (exit_DQ < 78) & (exit_DQ <= entry_DQ) & ((sub1_exit_rawScore > sub1_entry_rawScore) or (sub2_exit_rawScore > sub2_entry_rawScore)):
        return 'b'
    elif (exit_DQ < 78) & (exit_DQ > entry_DQ) & ((sub1_exit_rawScore > sub1_entry_rawScore) or (sub2_exit_rawScore > sub2_entry_rawScore)):
        return 'c'
    elif ((entry_DQ < 78) & (exit_DQ >= 78)) or (((entry_DQ >= 78) & (exit_DQ < 78) & ((sub1_ZScore <= -1.5) or (sub2_ZScore <= -1.5)))):
        return 'd'
    elif (exit_DQ >= 78) & (entry_DQ >= 78) :
        return 'e'
    else:
        return 'Fail'
```

```python
def get_category_3sub(exit_DQ, entry_DQ = 0, sub1_entry_rawScore=0, sub1_exit_rawScore=0, sub2_entry_rawScore=0, sub2_exit_rawScore=0, sub3_entry_rawScore=0, sub3_exit_rawScore=0, sub1_ZScore = -1.49, sub2_ZScore = -1.49, sub3_ZScore = -1.49):
    if (exit_DQ < 78) & (sub1_exit_rawScore <= sub1_entry_rawScore) & (sub2_exit_rawScore <= sub2_entry_rawScore) & (sub3_exit_rawScore <= sub3_entry_rawScore):
        return 'a'
    elif (exit_DQ < 78) & (exit_DQ <= entry_DQ) & ((sub1_exit_rawScore > sub1_entry_rawScore) or (sub2_exit_rawScore > sub2_entry_rawScore) or (sub3_exit_rawScore > sub3_entry_rawScore)):
        return 'b'
    elif (exit_DQ < 78) & (exit_DQ > entry_DQ) & ((sub1_exit_rawScore > sub1_entry_rawScore) or (sub2_exit_rawScore > sub2_entry_rawScore) or (sub3_exit_rawScore > sub3_entry_rawScore)):
        return 'c'
    elif ((entry_DQ < 78) & (exit_DQ >= 78)) or (((entry_DQ >= 78) & (exit_DQ < 78) & ((sub1_ZScore <= -1.5) or (sub2_ZScore <= -1.5) or (sub3_ZScore <= -1.5)))):
        return 'd'
    elif (exit_DQ >= 78) & (entry_DQ >= 78) :
        return 'e'
    else:
        return 'Fail'
```

# Ryan's Method

```python
def get_category_ryan(exit1_DQ, entry1_DQ = 0, exit2_DQ = 0, entry2_DQ = 0, sub1_entry_rawScore=0, sub1_exit_rawScore=0, sub2_entry_rawScore=0, sub2_exit_rawScore=0, sub3_entry_rawScore=0, sub3_exit_rawScore=0, sub4_entry_rawScore=0, sub4_exit_rawScore=0, sub5_entry_rawScore=0, sub5_exit_rawScore=0, sub1_ZScore = -1.49, sub2_ZScore = -1.49, sub3_ZScore = -1.49, sub4_ZScore = -1.49, sub5_ZScore = -1.49):
    if (exit1_DQ < 78) & (exit2_DQ < 78) & (sub1_exit_rawScore <= sub1_entry_rawScore) & (sub2_exit_rawScore <= sub2_entry_rawScore) & (sub3_exit_rawScore <= sub3_entry_rawScore) & (sub4_exit_rawScore <= sub4_entry_rawScore) & (sub5_exit_rawScore <= sub5_entry_rawScore):
        return 'a'
    elif (exit1_DQ < 78) & (exit1_DQ <= entry1_DQ) & (exit2_DQ < 78) & (exit2_DQ <= entry2_DQ) & ((sub1_exit_rawScore > sub1_entry_rawScore) or (sub2_exit_rawScore > sub2_entry_rawScore) or (sub3_exit_rawScore > sub3_entry_rawScore) or (sub4_exit_rawScore > sub4_entry_rawScore) or (sub5_exit_rawScore > sub5_entry_rawScore)):
        return 'b'
    elif (exit1_DQ < 78) & (exit1_DQ > entry1_DQ) & (exit2_DQ < 78) & (exit2_DQ > entry2_DQ) & ((sub1_exit_rawScore > sub1_entry_rawScore) or (sub2_exit_rawScore > sub2_entry_rawScore) or (sub3_exit_rawScore > sub3_entry_rawScore) or (sub4_exit_rawScore > sub4_entry_rawScore) or (sub5_exit_rawScore > sub5_entry_rawScore)):
        return 'c'
    elif (((entry1_DQ < 78) & (exit1_DQ >= 78)) & ((entry2_DQ < 78) & (exit2_DQ >= 78)) or (((entry1_DQ >= 78) & (exit1_DQ < 78) & (entry2_DQ >= 78) & (exit2_DQ < 78) & ((sub1_ZScore <= -1.5) or (sub2_ZScore <= -1.5) or (sub3_ZScore <= -1.5) or (sub4_ZScore <= -1.5) or (sub5_ZScore <= -1.55))))):
        return 'd'
    elif (exit1_DQ >= 78) & (entry1_DQ >= 78) & (exit2_DQ >= 78) & (entry2_DQ >= 78) :
        return 'e'
    else:
        return 'Fail'
```
